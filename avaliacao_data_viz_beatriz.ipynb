````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os 
import numpy as np

df = pd.DataFrame([
    ['Acidental', .09],
    ['Descuido', .24],
    ['Intencional', .01],
    ['Falso Alarme', .08],
    ['Infundada', .05],
    ['Indeterminada', .34],
    ['Curto Circuito', .19],
], columns=['causa', 'perc'])

df
````

````python
grouped_df = df.groupby('causa')['perc'].sum().sort_values(ascending=False)

sns.barplot(x=grouped_df.values, y=grouped_df.index, hue=grouped_df.index, palette="flare")
plt.xlabel('Percentagem')
plt.ylabel('Causa')
plt.title('Percentagem total por causa')
plt.grid(axis='x', linestyle='--', linewidth=0.5, alpha=0.5)
plt.show()
````

````python
df_heart = pd.read_csv(os.path.join('data', 'heart.csv'))
````

````python
grouped_df_sex = df_heart.groupby('sex')['target'].value_counts(normalize=True*100).rename('perc').reset_index()

grouped_df_sex['sex'] = np.where(grouped_df_sex['sex'] == 0, 'F', 'M')

sns.barplot(x= grouped_df_sex['sex'], y= grouped_df_sex['perc'], hue=grouped_df_sex['target'], palette=['lightgreen' , '#FF7F50'])
plt.xlabel('Percentagem')
plt.ylabel('Género')
plt.title('Percentagem de presença de doença cardíaca por género')
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.5)
plt.show()

grouped_df_sex 
````

````python
df_heart[['target', 'trestbps']].groupby('trestbps').sum().plot(style='-')
plt.xlabel('Pressão sanguínea em repouso')
plt.ylabel('Somatório de pessoas com doença cardíaca')
plt.title('Presença de doença cardíaca por nível de pressão sanguínea ')
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.5)
plt.grid(axis='x', linestyle='--', linewidth=0.5, alpha=0.5)
plt.show()
````

````python
stress_number = df_heart['thal'].value_counts().reset_index()
plt.pie(stress_number["count"], labels=stress_number["thal"], autopct='%1.1f%%')
sns.set_palette("husl")
plt.title('Percentagem de pessoas de acordo com o nível de sress')
plt.axis('equal')
sns.set_style("whitegrid")
plt.show()
stress_number
````

````python
grouped_df_age = df_heart.groupby('age')['target'].sum().reset_index()
plt.figure(figsize=(12, 6))
plt.xticks(rotation=45)
xticks = range(0, 101, 10)
sns.barplot(x= grouped_df_age['age'], y= grouped_df_age['target'], hue=grouped_df_age['target'], palette='Set2')
plt.xlabel('Idade')
plt.ylabel('Somatório de presença de doença cardíaca')
plt.title('Somatório de presença de doença cardíaca por idade')
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.5)
plt.show()

````
